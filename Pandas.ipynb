# Performing aggregation and grouping on data


```python
td = trend_data[-10:]
print(td)
```

```python
#now as the indexes are copied from previos dataframe 
# we create our new 
td = td.reset_index(drop=True)
td
```

```python
import numpy as np
```

```python
moving_average = td.rolling(window=3, min_periods=1)
moving_average.aggregate(np.min)
```

```python
# so now if we want to calculate the moving average
# rolling is great for time series analysis
# as the data on daily basis may fluctuate alot but eventually it is has same average as other days

moving_average = td.rolling(window = 3, min_periods=1)
moving_average['DataScience'].aggregate(np.mean)
```

```python
moving_average = td.rolling(window = 3, min_periods=1)
moving_average[['DataScience','MachineLearning','DeepLearning']].aggregate([np.mean,np.min,np.max])
```

## Group by

```python
trend_data
```

```python
#first we would add two new columns on Year and Month using the Column Week
trend_data['Year'] = trend_data['Week'].str.strip().str[-4:]
trend_data['Month'] = trend_data['Week'].str.strip().str[-7:-5]
trend_data.head()
```

```python
#so now we have the columns Year and Month in our data
trend_data.groupby('Year').DataScience.mean()
```

```python
trend_data.groupby('Year').apply(lambda trend_data: trend_data.mean())
```

```python
trend_data.groupby('Year').mean()
```

```python
trend_data.groupby(['Year','Month']).mean().head(10)
```

```python
trend_data.groupby(['Year','Month']).MachineLearning.mean().head(10)
```

```python
trend_data.groupby(['Year','Month']).aggregate({'DataScience':np.mean,'ArtificialIntelligence':np.mean})
```

```python
trend_data.groupby(['Year']).sum()
```

```python
seven_day_average = trend_data[trend_data['Year']>='2017'].rolling(window=7,min_periods=1)
seven_day_average['MachineLearning'].aggregate(np.mean).reset_index(drop=True)
```
